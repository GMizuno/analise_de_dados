<h1>Seaborn parte 4</h1>

Nessa parte vamos focar em plotar graficos que ajudam na hora de fazer um regressão linear

```python
import seaborn as sns
import numpy as np
%matplotlib inline
df = sns.load_dataset('tips')
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = df)
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = df, fit_reg = False) # Retira a banda de confiança e a reta de regressão.
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = df, hue = 'sex') # Criando categorias
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = df, hue = 'sex', palette = 'coolwarm')
# Mudando a cor
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = df, hue = 'sex', markers = ['o', 'v'])
# Mudando marcardor de cada grupo
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = df, hue = 'sex', markers = ['o', 'v'],
           scatter_kws = {'s': 100, 'alpha': 0.4})
# Altera padrões do marcardor, no nosso exemplo s = size e alpha = opacidade
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = df, hue = 'sex', col ='sex')
# Separando as regressões em colunas
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = df, hue = 'sex', row ='sex')
# Separando as regressões em linhas
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = df, hue = 'sex', col ='sex', row = 'time')
# Separando as regressões em linhas baseado na va time.
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = df, hue = 'sex', col ='day', aspect = 0.5, height = 8)
# Altera forma com os graficos estão dispostos, aspect = fornece a altura de cada quadrado (aspect = aspect * height)
# height = largura
```
