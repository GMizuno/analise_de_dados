<h1>Seaborn parte 4</h1>

Nessa parte vamos focar em plotar graficos que ajudam na hora de fazer um regress√£o linear

```python
import seaborn as sns
import numpy as np
%matplotlib inline
df = sns.load_dataset('tips')
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = df)
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = df, fit_reg = False) # Retira a banda de confian√ßa e a reta de regress√£o.
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = df, hue = 'sex') # Criando categorias
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = df, hue = 'sex', palette = 'coolwarm')
# Mudando a cor
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = df, hue = 'sex', markers = ['o', 'v'])
# Mudando marcardor de cada grupo
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = df, hue = 'sex', markers = ['o', 'v'],
           scatter_kws = {'s': 100, 'alpha': 0.4})
# Altera padr√µes do marcardor, no nosso exemplo s = size e alpha = opacidade
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = df, hue = 'sex', col ='sex')
# Separando as regress√µes em colunas
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = df, hue = 'sex', row ='sex')
# Separando as regress√µes em linhas
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = df, hue = 'sex', col ='sex', row = 'time')
# Separando as regress√µes em linhas baseado na va time.
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = df, hue = 'sex', col ='day', aspect = 0.5, height = 8)
# Altera forma com os graficos est√£o dispostos, aspect = fornece a altura de cada quadrado (aspect = aspect * height)
# height = largura
```
