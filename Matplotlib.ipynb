Para criar graficos com Matplotlib precisamos definir o espao ocupado pela figura (fig) e criar seus eixos (ax). O codigo acima faz de maneira curta, mas no intutiva, essa criao de maneira rapida. Agoa vou mostrar a forma 'mais lenta'.

```python
fig =  plt.figure(figsize = (12, 3)) # Criando figura sem eixo
fig = fig.add_subplot(1, 1, 1) # Coloquei eixos na figura
```

```python
fig =  plt.figure(figsize = (12, 3)) # Criando figura sem eixo
fig = fig.add_subplot(1, 1, 1) # Coloquei eixos na figura
fig.plot(x, y, 'r-')
fig.plot(x, 2*y, 'b-')
```

Apesar de esse exemplo trabalhar somente com uma figura no subplot  bem intuitivo o que deve ser feito. Basta fazer alterao nos parametros 3 do subplot.

```python
fig, ax = plt.subplots(nrows = 1, ncols = 2)
```

```python
ax
```

Temos duas formas de criar graficos quando estamos no cenario anterior.

Primeira forma, iterando.

```python
for axe in ax:
    axe.plot(x, y, 'b')
    axe.set_xlabel('Eixo X')
    axe.set_ylabel('Eixo Y')
    axe.set_title('Titulo')
fig
```

Segunda forma, acessando valores do array.

```python
ax[0].plot(x, y, 'b')
ax[0].set_xlabel('Eixo X')
ax[0].set_ylabel('Eixo Y')
ax[0].set_title('Titulo')

ax[1].plot(x, y, 'g')
ax[1].set_xlabel('Eixo X')
ax[1].set_ylabel('Eixo Y')
ax[1].set_title('Titulo')

fig
```

Agora supondo que queremos fazer um subplot com muitos graficos, com lidar com isso com as margens? Basta usar o mtodo **.tight_layout()** que  responsvel por 'ajeitar' as escalas.

```python
fig, ax = plt.subplots(nrows = 5, ncols = 5)
plt.tight_layout()
```

Para criar legenda basta passar os label em nos plot

```python
fig, axes = plt.subplots(figsize = (12, 3))
axes.plot(x, y, 'r-', label = 'x^2')
axes.plot(x, 2*y, 'g*--', label = 'Alguma coisa')
axes.legend()
```

Para mudar a posio da legenda basta passar o parametro loc, mas o Pyhton sempre tenta colocar na melhor possio.

```python
fig, axes = plt.subplots(figsize = (12, 3))
axes.plot(x, y, 'r-', label = 'x^2')
axes.plot(x, 2*y, 'g*--', label = 'Alguma coisa')
axes.legend(loc = 1)
```

<h2>Gerindo Layout</h2>

```python
fig, ax = plt.subplots(figsize = (8, 8))
ax.plot(x, x ** 2)
ax.plot(x, x ** 3)
```

```python
fig, ax = plt.subplots(figsize = (8, 8))
ax.plot(x, x ** 2, color = '#8B008B') # Hexadecimal
ax.plot(x, x ** 3, color = 'green') # string
```

```python
fig, ax = plt.subplots(figsize = (8, 8))
ax.plot(x, x ** 2, color = '#8B008B', linewidth = 1) # Espessura
ax.plot(x, x ** 3, color = 'green', lw = 10) # Espessura
```

```python
fig, ax = plt.subplots(figsize = (8, 8))
ax.plot(x, x ** 2, color = '#8B008B', alpha = 0.1) # Opacidade (entre 0 e 1)
ax.plot(x, x ** 3, color = 'green', alpha = 0.8) # Opacidade (entre 0 e 1)
```

```python
fig, ax = plt.subplots(figsize = (8, 8))
ax.plot(x, x ** 2, linestyle = '-')
ax.plot(x, x ** 3, linestyle = '--')
ax.plot(x, x ** 4, linestyle = ':')
```

```python
fig, ax = plt.subplots(figsize = (8, 8))
ax.plot(x, x ** 2, linestyle = '-', marker = '+') # Add marcador
ax.plot(x, x ** 3, linestyle = '-', marker = 'o') # Add marcador
ax.plot(x, x ** 4, linestyle = '-', marker = 's') # Add marcador
```

```python
fig, ax = plt.subplots(figsize = (8, 8))
ax.plot(x, x ** 2, marker = '+', markeredgecolor ='red', markerfacecolor ='blue') # Add cor de entrono e dentro do marcador
ax.plot(x, x ** 3, marker = 'o', markeredgecolor ='blue', markerfacecolor = 'red') # Add cor de entrono e dentro do marcador
ax.plot(x, x ** 4, marker = 's', markeredgecolor ='green', markerfacecolor = 'pink') # Add cor de entrono e dentro do marcador
```

```python
fig, ax = plt.subplots(figsize = (8, 8))
ax.plot(x, x ** 2)
ax.plot(x, x ** 3)

ax.set_xlim([0,4]) # Mudando limite do eixo X
ax.set_ylim([0,90]) # Mudando limite do eixo y
```
