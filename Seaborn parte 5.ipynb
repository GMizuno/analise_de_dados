<h2>Trabalhando com Grid</h2>

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
df = sns.load_dataset('iris')
```

```python
df.info()
```

```python
df.head()
```

```python
df['species'].value_counts()
```

O método PairGrid é o equivalente ao subplot.

```python
sns.PairGrid(df) # Criando subplot para cada variavel numerica do dataframe
```

Temos graficos vazios e para preencher eles usaremos a função **map()**.

```python
g = sns.PairGrid(df)
g.map(plt.scatter)
```

```python
g = sns.PairGrid(df)
g.map_diag(plt.hist) # Plotando somente na diagonal
```

```python
g = sns.PairGrid(df)
g.map_upper(plt.scatter) # Plotando somente na diagonal
```

```python
g = sns.PairGrid(df)
g.map_lower(sns.kdeplot)
```

Juntando map_diag, map_upper e map_lower

```python
g = sns.PairGrid(df)
g.map_diag(plt.hist)
g.map_upper(plt.scatter) 
g.map_lower(sns.kdeplot)
```

**Comparando sns.pairplot com sns.PairGrid()**

```python
sns.pairplot(df, hue = 'species')
```

O método **.FacetGrid** é o equivalente ao **.PairGrid** mas podemos passar variaveis categoricas.

```python
df = sns.load_dataset('tips')
```

```python
g = sns.FacetGrid(data = df, col = 'time', row = 'smoker')
```

```python
g = sns.FacetGrid(data = df, col = 'time', row = 'smoker')
g.map(plt.hist, 'total_bill') # O segundo parametro do g.map é o qual variavel vai para plt.hist
```

<h2>Gerindo Tamnaho do Grafico</h2>

```python
df = sns.load_dataset('tips')
```

```python
sns.set_style('darkgrid') # Muda o 'fundo' do grafico
sns.countplot(x = 'sex', data = df)
sns.despine()
```

```python
sns.set_style('white')
sns.countplot(x = 'sex', data = df)
sns.despine() # Retira a borda, por padrão as bordas de baixo e da direita ficam
```

```python
sns.set_style('ticks')
plt.figure(figsize = (12, 4)) # Alterando tamanho figura
sns.countplot(x = 'sex', data = df)
```

```python
sns.set_context('poster') # Alterando tamanho figura
sns.countplot(x = 'sex', data = df)
```

```python
sns.set_context('poster', font_scale = 2) # Alterando tamanho da fonte
sns.countplot(x = 'sex', data = df)
```
