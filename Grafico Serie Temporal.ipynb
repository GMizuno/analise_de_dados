```python
import datetime
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
np.random.seed(31)
```

Gerando os dias.

```python
datas = []
inicio = datetime.datetime(2019, 1, 1)
fim = datetime.datetime(2019, 1, 31)
delta = datetime.timedelta(days = 1)
for ii in range(1, 32):
    datas.append(inicio)
    inicio += delta
```

Gerando os valores de maneira artificial

```python
valores = np.random.randn(len(datas))
valores
```

```python
fig = plt.figure(figsize = (15,8))
np.random.seed(31)
sns.lineplot(x = datas, y = valores)
sns.despine()
```

Colocando duas series

```python
np.random.seed(31)
valores2 = np.random.normal(loc = 2, scale = 1, size = len(datas))
```

```python
fig, axes = plt.subplots(figsize = (15,8))
sns.set_style('darkgrid')
graf = sns.lineplot(x = datas, y = valores, label = 'Serie 1', color = 'blue')
graf = sns.lineplot(x = datas, y = valores2, label = 'Series 2', color = 'red')
axes.legend(loc='upper left', frameon=False)
sns.despine()
```

```python
datas = []
np.random.seed(31)
inicio = datetime.datetime(2019, 1, 1)
fim = datetime.datetime(2019, 1, 31)
delta = datetime.timedelta(days = 1)
for ii in range(1, 32):
    datas.append(inicio.strftime('%d/%b'))
    inicio += delta
valores = np.random.randn(len(datas))
fig, axes = plt.subplots(figsize = (20,8))
sns.lineplot(x = datas, y = valores, label = 'Serie 1')
sns.despine()
axes.set_title('Serie do ano de 2019')
```

```python
datas = []
np.random.seed(31)
inicio = datetime.datetime(2019, 1, 1)
fim = datetime.datetime(2019, 1, 31)
delta = datetime.timedelta(days = 1)

for ii in range(1, 32):
    datas.append(inicio.strftime('%d/%b'))
    inicio += delta
    
valores = np.random.randn(len(datas))

fig, axes = plt.subplots(figsize = (20,8))
sns.lineplot(x = datas, y = valores, label = 'Serie 1')
axes.set_title('Serie do ano de 2019')
axes.set_xticks(datas[0: 32: 2]) # Mudando o intervalo no eixo X.
```
