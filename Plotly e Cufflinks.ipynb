<h1>Plotly e Cufflinks</h1>

Nessa seção vamos trabalha com a bibliotecas Plotly e Cufflinks que são responsavéis por criar graficos interativos.

```python
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import cufflinks as cf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
init_notebook_mode(connected = True)
cf.go_offline()
```

```python
df1 = pd.DataFrame(np.random.randn(100, 4), columns = 'A B C D'.split())
df1.head()
```

```python
df2 = pd.DataFrame({'Categorias': ['A', 'B', 'C'], 'Valores': [32, 43, 50]})
df2
```

```python
df1.plot(kind = 'scatter', x = 'A', y = 'B', )
```

```python
df1.iplot(kind = 'scatter', x = 'A', y = 'B')
```

```python
df1.iplot(kind = 'scatter', x = 'A', y = 'B', mode = 'markers')
```

```python
df2.iplot(kind = 'bar', x = 'Categorias', y = 'Valores')
```

```python
df1.iplot(kind = 'bar')
```

Podemos usar metodos de agrupamento antes de usar o Plotly

```python
df1.count().iplot(kind = 'bar')
```

```python
df1.iplot(kind = 'box')
```

Trabalhando com superficies, para isso vamos usar Cufflinks

```python
df3 = pd.DataFrame({'x': [1, 2, 3, 4, 5], 'y': [10, 20, 30, 40, 50], 'z': [5, 4, 3, 2, 1]})
```

```python
df3.iplot(kind = 'surface')
```

```python
df3.iplot(kind = 'surface', colorscale = 'rdylbu') # Mudando escala de cores
```

```python
df1[['A', 'B']].iplot(kind = 'spread')
```

```python
df1['A'].iplot(kind = 'hist')
```

```python
df1['A'].iplot(kind = 'hist', bins = 50) # Mudando a quantidade de barras
```

Grafico de Bolhas.

É um grafico que avalia a dispressão com o tamanho dos pontos variando e função de outra variavel

```python
df1.iplot(kind = 'bubble', x = 'A', y = 'B', size = 'C')
```

Plotando grafico similar ao sns.pairplot

```python
df1.scatter_matrix()
```
