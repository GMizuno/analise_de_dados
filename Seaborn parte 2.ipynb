<h1>Seaborn parte 2</h1>

Na parte vamos focar em plotar 7 tipos de graficos que envolvem uma variavel numerica e outra categorica.

* Barplot -> Faz um grafico de barras.
* Countplot -> Faz um grafico de barras so que no eixo Y vai ter um contagem.
* Boxplot -> Faz um Boxplot.
* Violinplot -> Primo do Boxplot
* Stripplot -> Plot pontos.
* Swarmplot -> Junta o Stripplot com Violinplot

```python
import seaborn as sns
import numpy as np
%matplotlib inline
df = sns.load_dataset('tips')
```

```python
sns.barplot(x = 'sex', y = 'total_bill', data = df) # Por padrao no eixo Y vem a media
```

```python
sns.barplot(x = 'sex', y = 'total_bill', data = df, estimator = sum) # Mudando para soma
```

```python
sns.barplot(x = 'sex', y = 'total_bill', data = df, estimator = np.std) # Mudando para desvio padrao
```

```python
sns.countplot(x = 'sex', data = df)
```

```python
sns.boxplot(x = 'day', y = 'total_bill', data = df)
```

```python
sns.boxplot(x = 'day', y = 'total_bill', data = df, palette = 'rainbow') # Mudando cor
```

```python
sns.boxplot(x = 'day', y = 'total_bill', data = df, hue = 'sex') # Criando outro grupo  
```

```python
sns.boxplot(data = df, orient = 'h') # Girando o grafico
```

```python
sns.violinplot(x = 'day', y = 'tip', data = df)
```

```python
sns.violinplot(x = 'day', y = 'tip', data = df, hue = 'sex') # Criando graficos
```

```python
sns.violinplot(x = 'day', y = 'tip', data = df, hue = 'sex', split = True) 
# Como cada violino  simetrico o parametro split junta os grupo
```

```python
sns.stripplot(x = 'day', y = 'total_bill', data = df)
```

```python
sns.stripplot(x = 'day', y = 'total_bill', data = df, jitter = True) 
# Da um afastada nos ponto de um determinada catergoria
```

```python
sns.stripplot(x = 'day', y = 'total_bill', data = df, hue = 'sex') # Criando grupos
```

```python
sns.stripplot(x = 'day', y = 'total_bill', data = df, hue = 'sex', dodge = True) 
# Seperando os grupo craido no hue, o parametro dpdge funciona da mesma forma que split
```

```python
sns.swarmplot(x = 'day', y = 'total_bill', data = df)
```

```python
sns.swarmplot(x = 'day', y = 'total_bill', data = df, hue = 'sex') # Add grupo sex
```

```python
sns.swarmplot(x = 'day', y = 'total_bill', data = df, hue = 'sex', dodge = True) 
# Separando grupo criado em hue
```

```python
sns.swarmplot(x = 'day', y = 'total_bill', data = df, color = 'black')
sns.violinplot(x = 'day', y = 'total_bill', data = df)
```
