<h1>Seaborn parte 1</h1>

O Seaborn Ã© um biblioteca grafica, mas contem alguns datasets para testes.

Na parte vamos focar em plotar 5 tipos de graficos que envolvem variaveis numerica.

* Distplot -> Faz histograma junto com a densidade.
* Joinplot -> Plota um grafico de distribuicao conjunta.
* Pairplot -> Cria um subplot com as variaveis numericas e plota grafico de dispressao 2 a 2.
* Rugplot ->  https://en.wikipedia.org/wiki/Rug_plot
* Kdeplot -> https://pt.wikipedia.org/wiki/Estimativa_de_densidade_kernel

```python
import seaborn as sns
%matplotlib inline
```

```python
df = sns.load_dataset('tips')
```

```python
type(df)
```

```python
df.head()
```

```python
df.info()
```

```python
sns.distplot(df['total_bill'])
```

```python
sns.distplot(df['total_bill'], kde = False) # Tirando a curva
```

```python
sns.distplot(df['total_bill'], kde = False, bins = 50) # Alterando os intervalos no eixo X.
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df)
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind = 'reg') # Faze uma regerssao linear
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind = 'hex')
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue = 'sex') # Preenchendo por grupo
```

```python
sns.pairplot(df, hue = 'sex', palette = 'coolwarm') # Mudando paleta de cor das categorias
```

```python
sns.pairplot(df, palette = 'coolwarm') # Note q nao mudou nada
```

```python
sns.rugplot(df['total_bill'])
```
